### Question 15
Write a Python program to print the sum of the series 1/2+1/3+1/4+ ... +1/N. Where N is natural number.

```python
n = int(input("Enter value of N: "))
sum = 0

for i in range(2, n + 1):
    sum += 1/i

print("\nSum =", round(sum, 2))
```

### Question 16
Write a Python program that prompts user to enter numbers. The process will repeat until user enters 0. Finally, the program prints sum of the numbers entered by the user.

```python
sum = 0

while True:
    k = eval(input("Enter number: "))
    if k == 0: break
    sum += k

print("\nSum =", sum)
```

### Question 19
Write a Python program to check whether a number is Armstrong number or not.

```python
a = int(input("Enter number: "))
sum = 0
n = a

while n > 0:
    d = n % 10
    sum += d**3
    n //= 10

if a == sum: print("\nIt is an Armstrong number")
else: print("\nIt is not an Armstrong number")
```

### Question 20
Write a Python program to swap first and last digits of a number.

```python
n = int(input("Enter number: "))

l = list(str(n))
l[0], l[-1] = l[-1], l[0]

s = ""
for i in l: s += i
n = int(s)

print("\nNumber after swapping first and last digits:", n)
```

### Question 21
Write a Python program for printing prime numbers up to N. (N>100).

```python
def prime(n):
    flag = True

    if n > 1:
        for i in range(2, n):
            if n % i == 0: 
                flag = False
                break
            else: flag = True
    else: flag = False
    return flag

n = int(input("Enter number: "))

for i in range(1, n + 1):
    if prime(i): print(i, end=" ")
```
