### Question 1
Write a Python program to find the distance between two coordinate points (x1, y1) and (x2, y2).

```python
x1 = eval(input("Enter x1: "))
y1 = eval(input("Enter y1: "))
x2 = eval(input("Enter x2: "))
y2 = eval(input("Enter y2: "))

d = ((x2-x1)**2 + (y2-y1)**2)**0.5
print("\nDistance =", round(d, 2))
```

### Question 2
Write a Python program to input Percentage. Calculate percentage and grade according to following:

| Condition          | Output   |
|--------------------|----------|
| Percentage	>= 90% | Grade	A |
| Percentage	>= 80% | Grade	B |
| Percentage	>= 70% | Grade	C |
| Percentage	>= 60% | Grade	D |
| Percentage	>= 40% | Grade	E |
| Percentage < 40%   | Grade F  |

```python
p = eval(input("Enter percentage: "))

if p >= 90: g = "A"
elif p >= 80: g = "B"
elif p >= 70: g = "C"
elif p >= 60: g = "D"
elif p >= 40: g = "E"
else: g = "F"
    
print("Grade:", g)
```

### Question 4
Write a Python program that computes the real roots of a quadratic function. Your program should begin by prompting the user for the values of a, b and c. Then it should display a message indicating the nature of real roots, along with the values of the real roots (if any).

```python
a = eval(input("Enter a: "))
b = eval(input("Enter b: "))
c = eval(input("Enter c: "))

D = b**2 - 4*a*c
r1 = -b + D/(2*a)
r2 = -b - D/(2*a)

if D == 0: print(f"\nEqual roots: {r1} and {r2}")
elif D < 0: print(f"\nImaginary roots: {r1} and {r2}")
elif D > 0: print(f"\nReal roots: {r1} and {r2}")
```

### Question 5
Write a program to input angles of a triangle and check whether triangle is valid or not. Also, validate the angles entered by the user. (Sum of the three angles of triangle is 180)

```python
a = eval(input("Enter angle A: "))
b = eval(input("Enter angle B: "))
c = eval(input("Enter angle C: "))

if a > 0 and b > 0 and c > 0 and a + b + c == 180:
    print("\nValid triangle")
else:
    print("\nInvalid triangle")
    if a <= 0: print("\nAngle A is invalid")
    if b <= 0: print("\nAngle B is invalid")
    if c <= 0: print("\nAngle C is invalid")
```

### Question 6
Write a program to input basic salary of an employee and calculate its Gross salary according to following:

Basic	Salary	<=	10000	HRA	= 20%,	DA = 80%\
Basic	Salary	<=	20000	HRA = 25%,	DA = 90%\
Basic Salary > 20000 : HRA = 30%, DA = 95%



```python
BS = eval(input("Basic salary: "))

if BS <= 10000: HRA, DA = 20, 80
elif BS <= 20000: HRA, DA = 25, 90
else: HRA, DA = 30, 95
    
HRA, DA = BS*(HRA/100), BS*(DA/100)
GS = BS + HRA + DA
print("\nGross salary: ", GS)
```
