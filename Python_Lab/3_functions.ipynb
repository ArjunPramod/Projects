### Question 24
Define a function to find sum of all odd numbers between 1 to n.

```python
def sum_odd(n):
    sum = 0
    for i in range(1, n):
        if i % 2 != 0: sum += i
    return sum

n = int(input("Enter value of N: "))
print(f"\nSum of odd numbers between 1 to {n} = {sum_odd(10)}")
```

### Question 25
Define a function to check whether a number is palindrome or not.

```python
def palindrome(n):
    k = n
    rev = 0
    while(n > 0):
        d = n % 10
        rev = rev * 10 + d
        n //= 10
    if(k == rev): return True
    else: return False
        
n = int(input("Enter a number: "))

if palindrome(n): print("\nThe number is palindrome")
else: print("\nThe number isn't palindrome")
```

### Question 27
Define a function to check whether number is perfect or not.

```python
def perfect_no(n):
    sum = 0
    for i in range(1, n):
        if (n % i == 0): sum += i
    if(sum == n): return True
    else: return False
    
n = int(input("Enter number: "))

if perfect_no(n): print("\nIt is a perfect number")
else: print("\nIt is not a perfect number")
```

### Question 30
Define a recursive function to find power of a number.

```python
def power(n, p):
    if p == 0: return 1
    else: return n * power(n, p - 1)
    
n = eval(input("Enter number: "))
p = int(input("Enter power: "))
print(f"\n{n} to the power {p} = {power(n, p)}")
```

### Question 33

### Question 35
Write function for finding factors (n) and use factors function to check whether given number n is prime or not.

```python
def Factors(n):
    l = []
    c = 2
    while(n > 1):
        if(n % c == 0):
            l.append(c)
            n = n / c
        else: 
            c = c + 1
    return l

n = int(input("Enter number: "))
factors = Factors(n)

print(f"\nFactors of {n} = {factors}")

if len(factors) == 1: print("\nIt is a prime number")
else: print("\nIt isn't a prime number")
```

### Question 36
Write a python program for printing Fibonacci series
1. Write recursive approach implementation
2. Write iterative implementation

```python
# Recursive approach

def fibo_rec(n):
    if n <= 1: return n
    else: return fibo_rec(n-1) + fibo_rec(n-2)
    
n = int(input("Enter how many terms: "))
print("\nFibonacci sequence:", end=" ")

if n > 0:
    for i in range(10):
        print(fibo_rec(i), end=" ")
```

```python
# Iterative approach

def fibo_iter(n): 
    a, b = 0, 1
    for i in range(n):
        a, b = b, a + b
    return a

n = int(input("Enter how many terms: "))
print("\nFibonacci sequence:", end=" ")

if n > 0:
    for i in range(n):
        print(fibo_iter(i), end=" ")
```
